Our tool will use Twitter combined with weather data to help identify and report on likely locations of legitimate power outages using Natural Language Processing. Our metric to evaluate will be cosine similarity.

## Tweet Collection

With collecting tweets, we needed to decide if we were going to use a scraper or the Twitter api. The Twitter api looked promising at first because it included geolocation data. However, the api would only give us one month of data and we wanted historical tweets. So we shifted gears and tested a variety of Twitter scrapers. We settled on TwitterScraper because its functionality worked well for what we wanted to extract, historical data by location. While we wouldn’t be able to get geolocation from TwitterScraper, we would be able to narrow down tweets by city. 

To select the keywords, we spent time analyzing how people talked about power outages online, ranging from energy companies and news accounts to personal consumer accounts. We initially noticed that Texas was going through a power outage so started there with assessing how people were tweeting about it. One thing we noticed was the order of the words people used. And by changing one query to “power out” we were able to spot a lot more relevant outages. We ran tests using TwitterScraper to help us narrow down the search and to rule out words that would lead to high misclassification such as “blackout” because it’s a video game and uses of “power” that pulled in the tv show with the same name. 

For narrowing down region, we first tested some areas where power outages occur most frequently (based on Blackout Tracker’s annual reports [2](https://switchon.eaton.com/plug/blackout-tracker)). We settled on Michigan because of how consistently people were posting about power outages on Twitter. 

We tested the area by scanning tweets with our keywords in a radius around Detroit, but we noticed we were getting a lot of Canadian tweets too. So we narrowed our focus by pulling a city list we got from the Census and we targeted cities with a population of more than 10k. We ran a for loop to extract the city name from the list, then scanned for 15 miles around the city looking for use of our keywords during a certain date range. Here, we ran into a few problems. First, we ended up with strange results. Thousands of tweets from New Zealand appeared. We also got tweets from South Africa and Gainesville, FL. We thought we were able to fix this problem by making sure our search focused on the US only. However, the scanner kept getting stuck on certain cities and would take hours to scrape them, even if there were few tweets to pull. Because of the amount of time it took, two of us were locked out of scraping Twitter overnight and very few tweets were pulled.

We went back to the drawing board. We hypothesized that the scraper was having trouble with smaller cities. We decided to target the ten most populated cities in the US to see how that would perform.[3](https://www.moving.com/tips/the-top-10-largest-us-cities-by-population/) We got results quickly this way. We adjusted the cities we scraped slightly though. New York City pulled in quite a lot of noise and we didn’t want to train our model on that. We removed that city from our list. We added in the 11th top city instead (Austin) and also added in Detroit and Columbus because Michigan and Ohio rank high for yearly power outage rates via the Blackout Tracker reports. We scanned 5+ years worth of relevant tweets for each city (1/1/14 - 4/15/19) and used a radius of 25 miles. We planned on going back and testing a smaller radius here but ran out of time due to throttling issues. 

Regarding the throttling issues, we experienced these often and at times were locked out from scraping all together. We resolved this by bringing in more computers to scrape for us.

We saved an individual csv file for each city. This allowed us to spot any scraping issues along the way. It also helped with throttlings issues because we could easily see where the scrape left off.

### Web Scraping

What follows is the final version of the code for what we scraped on Twitter using TwitterScraper.

```python
#creating a list of cities we want to scrape
cities_list = ['Houston', 'Phoenix', 'San Antonio', 'San Jose', 'Philadelphia', 'Chicago', 'Austin', 'Dallas', 'San Diego', 'Los Angeles', 'Detroit', 'Columbus']
```

```python
#for loop to interate through our city list and scrape Twitter for a selection of targeted key words
for city in cities_list:
    df_tweets = pd.DataFrame(columns=['id','text','timestamp','user','location'])
    tweet_list = query_tweets(f'"outage" OR "power\'s out" OR "power out" OR "power outage" OR "pwr out" OR "power failure" OR "without power" OR "power is out" OR "electric is out" OR "electricity out" OR "electricity\'s out" OR "without electricity" OR "lost electricity" OR "lost power" OR "#poweroutage" -filter:retweets near:"{city}" within:10mi',
                              begindate = datetime.date(2014,1,1),
                              enddate = datetime.date(2019,4,15),
                              poolsize = 10)
    for row, tweet in enumerate(tweet_list):
        df_tweets.loc[row,'id'] = tweet.id
        df_tweets.loc[row,'text'] = tweet.text
        df_tweets.loc[row,'timestamp'] = tweet.timestamp
        df_tweets.loc[row,'user'] = tweet.user
        df_tweets.loc[row,'location'] = city
    df_tweets.to_csv(f'../datasets/tweets/{city}_tweets.csv')
```

## Weather Collection

From initial tweet review, we hypothesized that weather was the biggest influencer of a power outage. So we decided to scrape weather data to use along with our tweets. First, we needed to decide if we were using a scraper or API. We had hoped weather.gov’s API would work out best but the documentation wasn’t sufficient for figuring out how to scrape dated weather info. We reviewed a variety of API options though. Open Weather looked promising, however we could only get one month of data for free. We narrowed it down to two API options, a python wrapper for NOAA (National Oceanic and Atmospheric Administration) and Dark Sky. The NOAA wrapper had better functionality for what we were looking for, historical weather data by date for a specific region.

We pulled weather for our select cities based on the same date range we had pulled tweets for. We found a few interesting hiccups here that forced us to adjust what cities we’d be using. The NOAA wrapper uses a bounding box for latitude and longitudes to plot locations. We found an issue with San Antonio not having weather info available for all the years we were looking to pull. So we had to drop that city from our dataset too.

Another issue we found with collecting data is that each city has a different number of observation stations and each collects varying types of weather data. Some stations will collect different types of wind speed for example, while others will not. We honed in on the three types of info each station collected (high temp, low temp, and precipitation). Then we took an average of each city’s observation station data so that each day would have one reading available for each of the three weather readings. 

Each city's data was saved into an individual csv file, which made it easier for us to spot the differences between weather stations, as well as any scanning errors as they happened.

### Web Scraping

What follows is the final version of the code for what we pulled for weather via NOAA.

```python
#pulling in NOAA credentials - you will need to get an API key from NOAA for this part
#we saved key into a separate file to run
with open('./noaa_credentials.json') as cred_data:
    info = json.load(cred_data)
    noaa_cred = info['token']
```

```python
#dictionary of latitudes and longitudes
city_dict = {'austin': ['30.0968', '29.6019', '-96.0042', '-96.6221'],
             'chicago': ['42.0229', '41.6446', '-87.5245', '-87.9395'],
             'columbus': ['40.1392', '39.8101', '-82.7713', '-83.1838'],
             'dallas': ['33.0164', '32.6096', '-96.5396', '-97.0000'], 
             'detroit': ['42.4502', '42.2555', '-82.9105', '-83.2877'], 
             'houston': ['31.5928', '30.9261', '-94.9577', '-95.7762'],
             'philadelphia': ['40.1380', '39.8670', '-74.9558', '-75.2803'],
             'los_angeles': ['34.8233', '32.8007', '-117.6462', '-118.9448'],
             'phoenix': ['33.8927', '33.5583', '-111.9253', '-112.3240'],
             'san_diego': ['33.5053', '33.0198', '-116.0810', '-117.5962'],
             'san_jose': ['37.4692', '37.1255', '-121.5891', '-122.0457']}
```

```python
#looping through the ditionary to pull lat/long info
for city in city_dict:
    # initialize a client with a developer token ,
    # note 5 calls per second and 1000 calls per day limit for each token
    token = noaa_cred
    my_client = Client(token, default_units=None, default_limit=1000)
    
    # define the extent of each city using their bounding box
    extent = {
        "north": city_dict[city][0],
        "south": city_dict[city][1],
        "east":  city_dict[city][2],
        "west":  city_dict[city][3]
    }

    # setting the date range
    startdate = datetime(2014, 1, 1)
    enddate   = datetime(2019, 4, 15)

    # after examining the available datasets, we decided 'GHCND' is the one we want,
    # and that we really want daily min and max temperatures
    datasetid='GHCND'
    datatypeid=['TMAX','PRCP']

    # finding stations that meet all our criteria
    stations = my_client.find_stations(
        datasetid=datasetid,
        extent=extent,
        startdate=startdate,
        enddate=enddate,
        datatypeid=datatypeid,
        return_dataframe=True)

    # we can get big lists of station data with
    big_df = pd.DataFrame()
    for rowid, station in stations.iterrows():  
        station_data = my_client.get_data_by_station(
            datasetid=datasetid,
            stationid=station['id'],    
            startdate=startdate,
            enddate=enddate,
            return_dataframe=True,  
            include_station_meta=True   
        )

        big_df = pd.concat([big_df, station_data], sort=False)

    # Sorting the df by date and saving it
    big_df = big_df.sort_values(by='date').reset_index()
    big_df.to_csv(f'../datasets/weather/{city}_ghcnd_Jan-1-14_Apr-15-19.csv')
```

```python
# Pulling the city csvs into one a dictionary
noaa_city_dict = {}
for city in city_dict:
    noaa_city_dict[city] = {}
    noaa_city_dict[city][f'{city}_raw'] = pd.read_csv(f'../datasets/weather/{city}_ghcnd_Jan-1-14_Apr-15-19.csv')
    noaa_city_dict[city][f'{city}_raw'].drop(columns=['Unnamed: 0','index'], inplace=True)
    noaa_city_dict[city][f'{city}_raw']['date'] = pd.to_datetime(noaa_city_dict[city][f'{city}_raw']['date']).dt.date
```

```python
#Verifying that the dataframe was set up as intended
noaa_city_dict['austin']['austin_raw'].head()
```

Because we are using Natural Language Processing, we have to convert the numbers into words.  For the max and min temperature data, we labeled each based on what temperature range it fell in.[4](https://en.wikipedia.org/wiki/Orders_of_magnitude_(temperature)#Detailed_list_for_100_K_to_1000_K) For precipitation we labeled each based on rain intensity.[5](https://en.wikipedia.org/wiki/Rain#Intensity)

```python
#creating a dictionary of weather labels
data_dict = {
    'PRCP': {'bins':[-np.inf, 0.1, 2.51, 10, 50.01, np.inf],
             'names':['dry', 'drizzle', 'rain', 'downpour', 'torrential']},
    'TMAX': {'bins':[-np.inf,0.01,100.00,210.00,328.00,np.inf],
             'names':['freezing','cold','warm','hot','scorching']},
    'TMIN': {'bins':[-np.inf,0.01,100.00,210.00,328.00,np.inf],
             'names':['freezing ||','cold ||','warm ||','hot ||','scorching ||']}
}
```

```python
#creating a for loop to calculate the means for max temp, min temp, and precipitation based on number
#of observation stations per city
for city in noaa_city_dict:
    noaa_city_dict[city][f'{city}_mean'] = noaa_city_dict[city][f'{city}_raw'].groupby(by = 'date',as_index=False).mean()
    
    cityname = ''
    for name in city.split('_'):
        cityname += name.capitalize()
        cityname += ' '
    noaa_city_dict[city][f'{city}_mean']['city'] = cityname.strip()
    noaa_city_dict[city][f'{city}_mean']['date_place'] = noaa_city_dict[city][f'{city}_mean'][['date','city']].apply(lambda row: ', '.join(row.values.astype(str)), axis=1)
    print(f'City: {cityname}')
    print(noaa_city_dict[city][f'{city}_mean'].head())
    text_cols = []
    for feat in data_dict:
        text_cols.append(f'{feat}_text')
        bins = data_dict[feat]['bins']
        names = data_dict[feat]['names']
        noaa_city_dict[city][f'{city}_mean'][f'{feat}_text'] = pd.cut(noaa_city_dict[city][f'{city}_mean'][f'{feat}'], bins, labels=names)
    noaa_city_dict[city][f'{city}_mean']['combined_text'] = noaa_city_dict[city][f'{city}_mean'][text_cols].apply(lambda row: ', '.join(row.values.astype(str)), axis=1)
    print(noaa_city_dict[city][f'{city}_mean']['combined_text'].head())
```

## Exploratory Data Analysis

For EDA we wanted to take a look at what the overall story was that our data was trying to tell us. We decided to check out what the most popular words were that we scraped. In order to do that, we had to clean the data by lower casing it and removing punctuation before we could run it through CountVectorizer to get the word counts.

What we first discovered was that there was a lot of common terms getting in the way of understanding what the data was telling us. There were also pieces of urls. Plus, Instagram and Twitter were mentioned a lot. So we removed these words in order to see the main topics being discussed more clearly. We removed them as part of a custom list where we added words to it to remove as we reviewed and explored what words appeared. It helped with understanding more clearly what the language patterns were that we'd need to pay attention to when processing for the model.

```python
#reading in the file
df = pd.read_csv('../datasets/combined_data.csv')
```

```python
#cleaning dataframe text column
def text_clean(text):
       
    # removing non-letters
    letters_only = re.sub("[^a-zA-Z]", " ", text)
    
    # converting to lower case, split into individual words
    words = letters_only.lower().split()
        
    # creating a custom list of stopwords then converting it to a set to run faster
    stop_words = ['dlvr', 'us', 'http', 'www', 'https', 'com', 'ourselves', 'hers', 'between', 
                  'yourself', 'but', 'again', 'there', 'about', 'once', 'during', 'out', 
                  'very', 'having', 'with', 'they', 'own', 'an', 'be', 'some', 'for', 'do', 
                  'its', 'yours', 'such', 'into', 'of', 'most', 'itself', 'other', 'off', 
                  'is', 's', 'am', 'or', 'who', 'as', 'from', 'him', 'each', 'the', 
                  'themselves', 'until', 'below', 'are', 'we', 'these', 'your', 'his', 
                  'through', 'don', 'nor', 'me', 'were', 'her', 'more', 'himself', 'this', 
                  'down', 'should', 'our', 'their', 'while', 'above', 'both', 'up', 'to', 
                  'ours', 'had', 'she', 'all', 'no', 'when', 'at', 'any', 'before', 'them', 
                  'same', 'and', 'been', 'have', 'in', 'will', 'on', 'does', 'yourselves', 
                  'then', 'that', 'because', 'what', 'over', 'why', 'so', 'can', 'did', 
                  'not', 'now', 'under', 'he', 'you', 'herself', 'has', 'just', 'where', 
                  'too', 'only', 'myself', 'which', 'those', 'i', 'after', 'few', 'whom', 
                  't', 'being', 'if', 'theirs', 'my', 'against', 'a', 'by', 'doing', 'it', 
                  'how', 'further', 'was', 'here', 'than', 'twitter', 'instagram', 'node', 
                  'toplocalnow', 'cityandpress', 'pic', 'bit', 'ly', 'utm', 'source',
                 'ig', 'share', 'igshid']
    stops = set(stop_words)
    
    # removing stop words
    meaningful_words = [w for w in words if not w in stops]
    
    # joining the words back into one string separated by space, 
    # and return the result.
    return(" ".join(meaningful_words))
```

```python
#initializing an empty list to hold the clean posts
clean_text = []
```

```python
#setting up a for loop to iterate through the column
j = 0
for text_update in df['text']:
    clean_text.append(text_clean(text_update)) 
    j += 1
```

```python
#pulling the list values into a column
df['clean text'] = clean_text
```

```python
#dropping the unnamed column
df.drop(columns='Unnamed: 0', inplace = True)
#verifying that the text was cleaned
df.head(1)
```

### Single-Word Term Counts

We ran CountVectorizer to find out what the most used words were.

```python
#instantiate
cvec = CountVectorizer()
```

```python
#extracting a top range of words based on word count
top_words = pd.DataFrame(cvec.fit_transform(df['clean text']).todense(),
                             columns = cvec.get_feature_names())

top_words.sum()[top_words.sum() >= 200].sort_values(ascending = False)
```

```python
#setting a value for just the top ten words
top10 = top_words.sum().sort_values(ascending = False)[:10]
```

```python
#plotting frequency of words
plt.style.use('bmh')
plt.figure(figsize = (12,12))
top10.sort_values(ascending=True).plot(kind = "barh", color="#13addd")
plt.title("Top 10 Single-Word Terms in Tweets", fontsize=28,fontweight='bold')
plt.xlabel("Frequency Count", fontsize=26,fontweight='bold')
plt.ylabel("Term", fontsize=24,fontweight='bold')
plt.xticks(fontsize= 22,fontweight='bold')
plt.yticks(fontsize= 22,fontweight='bold')
plt.show();
```

Looking at the top words, we were not surprised to see "power", "outage", and "without" at the top. Those were among our targeted search terms when we scraped the data. What's intersting though is seeing "customer" and "area" which are terms used on energy provider accounts a lot. "Still" and "time" make sense as well since they refer to customers still being withouth power. Also interesting to see "Detroit" listed because Michigan is one of the top states based on yearly outage rates. 

### Multi-Word Term Counts

We thought that single-word terms didn't tell enough of the story though and decided to pull the top two and three word combos.

```python
#seeing if adding two word terms tell a better story about the top words
cvec = CountVectorizer(ngram_range=(2, 3))
```

```python
#calling out the top terms
top_words = pd.DataFrame(cvec.fit_transform(df['clean text']).todense(),
                             columns = cvec.get_feature_names())

top_words.sum()[top_words.sum() >= 75].sort_values(ascending = False)
```

```python
#setting a value for just the top ten words
top10 = top_words.sum().sort_values(ascending = False)[:10]
```

```python
#plotting frequency of words
plt.style.use('bmh')
plt.figure(figsize = (12,12))
top10.sort_values(ascending=True).plot(kind = "barh", color="#13addd")
plt.title("Top 10 Multi-Word Terms in Tweets", fontsize=28,fontweight='bold')
plt.xlabel("Frequency Count", fontsize=26,fontweight='bold')
plt.ylabel("Term", fontsize=24,fontweight='bold')
plt.xticks(fontsize= 22,fontweight='bold')
plt.yticks(fontsize= 22,fontweight='bold')
plt.show();
```

Looking at the top multi-word terms, we can see a more complete picture of the overal story of our dataset. Energy companies have customers without power. This is not a power outage and we will keep this in mind as we move forward with our model and with how we look at classifying those tweets that are not legitimate power outages.
